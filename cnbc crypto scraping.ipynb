```python
from bs4 import BeautifulSoup
import requests
import time
import datetime
import smtplib
import pandas as pd
import warnings
import json

# Ignore all warnings
warnings.filterwarnings("ignore")

#creating a session
session = requests.Session()
```

```python
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36"}
url = "https://www.cnbc.com/cryptoworld/"
page = session.get(url, headers=headers)
soup = BeautifulSoup(page.content, 'html.parser')

# getting the links for data collection
a=[]
my_links = soup.select('a.Card-title')
for link in my_links:
    link_url = link['href']
    a.append(link_url)

#initialising the lists for storing data
headline_lst = []
ext_news_lst = []

# getting and storing data
for link in a:
    # get the URL of the linked page
    link_url = link

    # use requests to retrieve the HTML content of the linked page
    linked_page = requests.get(link_url,headers=headers)

    # create a BeautifulSoup object from the HTML content of the linked page
    linked_soup = BeautifulSoup(linked_page.content, 'html.parser')

    # find the headline with the desired class and get its text
    headline = linked_soup.find('h1' , class_='ClipPlayer-clipPlayerIntroTitle')
    if(headline is not None):
        headline_lst.append(headline.text)
    else:
        headline_lst.append("")
    
    
    # find the extended news with the desired class and get its text
    ext_news = linked_soup.find('div' , class_='ClipPlayer-clipPlayerIntroSummary')
    if(ext_news is not None):
        ext_news_lst.append(ext_news.text)
    else:
        ext_news_lst.append("")
        
# Create a dictionary from the lists
data = {'Headline': headline_lst, 'News': ext_news_lst}

# Create a dataframe from the dictionary
crypto_news = pd.DataFrame(data)
crypto_news
```
